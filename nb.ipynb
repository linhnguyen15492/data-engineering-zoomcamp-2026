```python
sql = """
select revenue_monthly_total_amount, revenue_month, extract(year from revenue_month) as year, pickup_zone      
from prod.fct_monthly_zone_revenue
where service_type = 'Green' and extract(year from revenue_month) = 2020
order by revenue_monthly_total_amount desc
"""
```

```python
with duckdb.connect("taxi_rides_ny/taxi_rides_ny.duckdb") as conn:
    conn.sql(sql).show()
```
